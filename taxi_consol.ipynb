```python
import pandas as pd
```

```python
taxis_sub = pd.read_csv('data/Chicago/cleaned/taxis/taxis_count_Q1.csv')
```

```python
taxis_sub1 = pd.read_csv('data/Chicago/cleaned/taxis/taxis_count_Q2.csv')
```

```python
taxis_consol = pd.concat([taxis_sub, taxis_sub1]).groupby(['from_zip', 'to_zip']).sum().reset_index()
taxis_consol
```

```python
taxis_sub1 = pd.read_csv('data/Chicago/cleaned/taxis/taxis_count_Q4.csv')
taxis_consol = pd.concat([taxis_consol, taxis_sub1]).groupby(['from_zip', 'to_zip']).sum().reset_index()
taxis_consol
```

```python
taxis_consol.to_csv('data/Chicago/cleaned/taxis/taxis_2022.csv', index=False)
```
