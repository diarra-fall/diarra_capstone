```python
import pandas as pd
```

```python
rs_sub = pd.read_csv('data/Chicago/cleaned/rideshare/rs_count_m1.csv')
```

```python
rs_sub1 = pd.read_csv('data/Chicago/cleaned/rideshare/rs_count_m2.csv')
```

```python
rs_consol = pd.concat([rs_sub, rs_sub1]).groupby(['from_zip', 'to_zip']).sum().reset_index()
rs_consol
```

```python
rs_sub1 = pd.read_csv('data/Chicago/cleaned/rideshare/rs_count_m6.csv')
rs_consol = pd.concat([rs_consol, rs_sub1]).groupby(['from_zip', 'to_zip']).sum().reset_index()
rs_consol
```

```python
rs_consol.to_csv('data/Chicago/cleaned/rideshare/rs_2022.csv', index=False)
```
